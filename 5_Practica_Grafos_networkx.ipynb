````python
import networkx as nx
import matplotlib.pyplot as plt

def plot_graph(G):
    """
    funcion auxiliar para imprimir los grafos con los que trabajemos
    El peso de las aristas, si los hay, debe estar en el atributo 'weight'
    """
    pos=nx.spring_layout(G)
    nx.draw_networkx_nodes(G,pos)
    edge_labels=dict([((u,v,),d.get('weight', ''))
             for u,v,d in G.edges(data=True)])
    nx.draw_networkx_labels(G,pos, font_size=20,font_family='sans-serif')
    nx.draw_networkx_edges(G,pos)
    plt.axis('off')
    #nx.draw_networkx_labels(self.G,pos, font_size=20,font_family='sans-serif')
    nx.draw_networkx_edge_labels(G,pos,edge_labels=edge_labels)
    nx.draw(G,pos, edge_cmap=plt.cm.Reds)

    plt.show()
````

````python
from math import sqrt

class Grid:
  """
  Clase Grid: Representa una grilla de tamañio ancho x alto a partir de un string
  """
  def __init__(self, string, ancho, alto):
    """
    Construye una nueva instancia de la clase Grid, basado en un string, dado el ancho y alto
    """
    raw_split = string.replace("\n", " ").replace("\t", " ").split(" ")

    self.cells = [symbol for symbol in raw_split if symbol != ""]
    self.ancho = ancho
    self.alto = alto

  def at(self, x, y):
    """
    Devuelve el valor en la posición (x,y) en orden de lectura
    x crece hacia la derecha
    y crece hacia abajo
    """
    assert 0 <= x < self.ancho
    assert 0 <= y < self.alto
    return self.cells[x + self.ancho * y]
````
